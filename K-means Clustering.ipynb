# K-means Clustering
K means clustering은 n개의 객체를 특정 K개의 부분으로 그룹화 하는 것이다. (이때, n > k)

# 목차
    0. K-means clustering steps
    1. 데이터 탐색
    2. K-Means Clustering 
    3. Select Optimal number of K (최적의 K값 결정)

--------
## 0. K-means clustering Steps

1. Decide the value of k = number of clusters.

2. Randomly guess k cluster “center” locations.
    (means select random k objects as cluster centers)

3. Each object finds out which cluster center it is closest to.
    Then, each cluster owns a set of objects.

4. Each cluster finds the center of the objects it owns.

5. Jump to step 3, if cluster centers change. Otherwise, terminate.

--------

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# 1. 데이터 탐색
해당 데이터는 3000개의 2차원 벡터 값이다.

```python
df = pd.read_csv('./data/data_week11.csv',encoding ='utf-8')
df
```

# 2. K-Means Clustering
사이킷런(sklearn) 라이브러리를 이용해 K-means clustering을 진행한다.    
사이킷런 - KMeans의 parameter로 K값(clustering의 개수)을 지정하면, K means clustering을 실행한다.   
Kmeans의 함수인 cluster_centers_를 통해 해당 K개 만큼의 중심 좌표를 얻을 수 있다.  
이렇게 얻은 cluster_centers_를 출력하고, plt를 이용해 df의 데이터 및 중심 좌표를 출력한다.     

## K 개수에 따른 K-means clustering 그래프 확인  (2개, 3개, 4개, 6개, 9개, 12개) 

```python
k = 2
kmeans = KMeans(n_clusters=k).fit(df)
centroids = kmeans.cluster_centers_

for i,center in zip(range(1,k+1),centroids):
    print('중심 좌표',i, center)

plt.scatter(df['V1'], df['V2'], c= kmeans.labels_.astype(float), s=50, alpha=0.5) # s= marker size , alpha= 투명도
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

# 3. Select Optimal number of K (최적의 K값 결정)
최적의 K값을 결정하기 위해 Elbow method를 사용한다.
  
아래는 Elbow method에 대한 Wiki를 인용했다.  
In cluster analysis, the elbow method is a heuristic used in determining the number of clusters in a data set.   
The method consists of plotting the explained variation as a function of the number of clusters,    
and picking the elbow of the curve as the number of clusters to use.

위의 설명에서 알 수 있듯이 Elbow method는 Clustering에서 안정적인 클러스터의 개수를 선택하는 방법이다.  
  

<img style="float: left;" src="https://raw.githubusercontent.com/heonsooo/Data-Analytics/main/data/P_function.jpg"> 
왼쪽의 P 함수에서     
xi : j cluster 안에 있는 object    
cj : j culster내의 중심 좌표  
dist(xi, cj) : xi와 cj의 거리   
따라서 위의 함수는 cluster내의 중심 좌표로부터 각 개체(data)의 거리를 총합한 것이 P값이다.  
(이때, P값이 가장 작을때, 각 군집내에서의 거리 총합이 가장 클러스터의 개수를 의미한다.)
  
하지만, Cluster의 개수가 너무 많으면 데이터가 과도하게 세분화 된다.    
따라서 Elbow method에서는 최소의 P값을 갖는 K값이 아닌 최적의(거리의 합이 안정화 되는) K값을 찾는 것이다.

이를 통해 Elbow Method를 이용하여 적절한 elbow point를 선택하여 최적의 클러스터 개수 K를 결정할 것이다.

## 3-1 K개수 ( 1 ~ 30개) 


```python
inertias = []
num_clusters = list(range(1,31))

for i in num_clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(15,10))
plt.plot(num_clusters, inertias,marker='o' ,markersize=10)
plt.xlabel('Number of Clusters',  fontsize=18)
plt.ylabel('Sum of Distances ',  fontsize=18)
plt.show()
```

이 그래프에서 볼 수 있는 것처럼   
Cluster값이 0과 5사이의 특정 값에서 부터 stablize되는 것을 확인할 수 있다.  
이를 자세히 확인하기 위해 Cluster개수 K의 최댓값을 10으로 설정 후 그래프로 표현하면 다음과 같다.  

## 3-2 K개수 (2 ~ 12개)

```python
inertias = []
num_clusters = list(range(1,13))
for i in num_clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(15,10))
plt.plot(num_clusters, inertias,marker='o' ,markersize=10)
plt.xlabel('Number of Clusters',  fontsize=18)
plt.ylabel('Sum of Distances ',  fontsize=18)
plt.xticks(num_clusters)
plt.show()
```

### K=3 이후에 stabllize 된 현상을 발견
위의 그래프를 통해 distance의 합이 적절히 꺾이는 최적의 지점(Elbow Point)으로 K =3이 가장 적절한 클러스터의 개수라고 판단된다.

3이상의 클러스터 개수에서도 감소 추세를 더 세분화하여 파악하기 위해   
K값을 3부터 30으로 설정하여 다시한번 진행해 보았다.

## 3-3 K개수 (3 ~ 30개)

```python
inertias = []
num_clusters = list(range(3,31))
for i in num_clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
plt.figure(figsize=(15,10))
plt.plot(num_clusters, inertias,marker='o' ,markersize=10)
plt.xlabel('Number of Clusters',  fontsize=18)
plt.ylabel('Sum of Distances ',  fontsize=18)
plt.xticks(num_clusters)
plt.show()
```

위의 그래프에서 클러스터의 개수가 9일때 꺾이는 추세를 발견할 수 있다. 

하지만, 위의 그래프(K : 1~ 12개)의 경우보다 stabilize가 덜한 것을 확인할 수 있으며,   
 클러스터 개수가 1이 아닌 3부터 시작했다는 것에 문제가 있다고 생각한다.  

따라서 K의 값이 1부터 시작하는 K : 1 ~ 12개의 그래프에서 확인한 K = 3일때 가장 최적의 클러스터 개수라고 결정했다.

## 결론 : Optimal number of  K is 3


```python
k = 3
kmeans = KMeans(n_clusters=k).fit(df)
centroids = kmeans.cluster_centers_

for i,center in zip(range(1,k+1),centroids):
    print('중심 좌표',i, center)

plt.scatter(df['V1'], df['V2'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```
